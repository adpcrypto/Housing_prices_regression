## Importing Libraries 

```python
import pandas as pd
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt
import math
from sklearn.linear_model import LinearRegression

#display all the columns of the dataset
pd.pandas.set_option("display.max_columns", None)
```

```python
dataset = pd.read_csv("test.csv")
dataset.head(10)
```

## Handling missing data

```python
#returns all categorical features with missing values
cat_features_na = [features for features in dataset.columns if dataset[features].isnull().sum()>1 and dataset[features].dtype=='O']

for feature in cat_features_na:
    print(feature,": ", np.round(dataset[feature].isnull().mean(), 3)*100, "% missing")
```

```python
dataset.head()
```

```python
#Dropping Features with a lot of missing values
dataset.drop(['Alley', 'PoolQC', 'Fence', 'MiscFeature'], axis=1, inplace=True)
```

```python
#Replacing categorical values with mode 
cat_features_na = [features for features in dataset.columns if dataset[features].isnull().sum()>1 and dataset[features].dtype=='O']

for feature in cat_features_na:
    dataset[feature] = dataset[feature].fillna(dataset[feature].mode()[0])
```

```python
for feature in cat_features_na:
    print(feature,": ", np.round(dataset[feature].isnull().mean(), 3)*100, "% missing")
```

```python
#Replacing numerical values with missing data

num_features_na = [features for features in dataset.columns if dataset[features].isnull().sum()>1 and dataset[features].dtype!='O' ]

for feature in num_features_na:
    print(feature,": ", np.round(dataset[feature].isnull().mean(), 3)*100, "% missing")
```

```python
for feature in num_features_na:
    median = dataset[feature].median()
    
    dataset[feature].fillna(median, inplace=True)

dataset[num_features_na].isnull().sum()
```

```python
#Replacing temporal values

for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
    dataset[feature] = dataset['YrSold'] - dataset[feature]

dataset.head()
```

```python
numerical_feat = [feature for feature in dataset.columns if dataset[feature].dtypes != 'O']
categorical_feat=[feature for feature in dataset.columns if dataset[feature].dtypes=='O']

len(numerical_feat), len(categorical_feat)
```

```python
numerical_feat
```

## Scaling numerical Values

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(dataset[numerical_feat])
dataset[numerical_feat] = scaler.transform(dataset[numerical_feat])

dataset.shape

dataset.to_csv("X_test_unen.csv", index=False)
```

## Encode Categorical Values

```python
from sklearn.preprocessing import OneHotEncoder
# 1. Create the encoder
encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')
# 2. Fit the encoder to the categorical colums
encoder.fit(dataset[categorical_feat])
```

```python
encoded_cols = list(encoder.get_feature_names(categorical_feat))
encoded_cols
```

```python
dataset[encoded_cols] = encoder.transform(dataset[categorical_feat])
dataset.head()
```

```python
inputs = dataset.drop(categorical_feat, axis=1)
inputs.head()
```
